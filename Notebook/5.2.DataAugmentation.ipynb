```python
data = pd.read_csv('../Data/3.normalized_dataset_image.csv')
path = '../Data/Flipkart/Images/'
data
```

```python
import cv2
import random
import shutil
import os

source_file = '../Data/Flipkart/Images/'
goal_file = '../Data/AugmentedImages/'

df = []

def save_img(source_file, goal_file, img):
    source_path = os.path.join(source_file, img)
    goal_path = os.path.join(goal_file, img)

    if os.path.exists(source_path):
        shutil.copyfile(source_path, goal_path)
    else:
        print(f'error : {img} does not exist')

def augmented1(image, pathJPG, label):
    angle = np.random.uniform(-30, 30)
    rows, cols, _ = image.shape
    rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    augmented_image = cv2.warpAffine(image, rotation_matrix, (cols, rows)) # rotation
    output_path = goal_file+'AI'+pathJPG
    cv2.imwrite(output_path, augmented_image)
    df.append({'image_path' : 'AI'+pathJPG, 'label' : label})

def augmented2(image, pathJPG, label):
    tx = np.random.uniform(-150, 150)
    ty = np.random.uniform(-150, 150)
    rows, cols, _ = image.shape
    translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
    augmented_image = cv2.warpAffine(image, translation_matrix, (cols, rows)) #translation
    output_path = goal_file+'AI'+pathJPG
    cv2.imwrite(output_path, augmented_image)
    df.append({'image_path' : 'AI'+pathJPG, 'label' : label})

def augmented3(image, pathJPG, label):
    y=0
    x=0
    h=int((image.shape[1])/2)
    w=int((image.shape[0])/2)
    augmented_image = image[x:w, y:h] # crop
    output_path = goal_file+'AI'+pathJPG
    cv2.imwrite(output_path, augmented_image)
    df.append({'image_path' : 'AI'+pathJPG, 'label' : label})

def augmented4(image, pathJPG, label):

    augmented_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Conversion en format RGB
    output_path = goal_file+'AI'+pathJPG
    cv2.imwrite(output_path, augmented_image)
    df.append({'image_path' : 'AI'+pathJPG, 'label' : label})

def random_augmentedImage(image, pathJPG, label):
    r = random.randint(0,3)
    if r == 0:
        augmented1(image, pathJPG, label)
    elif r == 1:
        augmented2(image, pathJPG, label)
    elif r == 2:
        augmented3(image, pathJPG, label)
    elif r == 3:
        augmented4(image, pathJPG, label)



for el in data.iterrows():
    save_img(source_file, goal_file, el[1]['image_path'])

    df.append({'image_path' : el[1]['image_path'], 'label' : el[1]["label"]})

    image_path = f'../Data/Flipkart/Images/{el[1]["image_path"]}'
    original_image = cv2.imread(image_path)
    random_augmentedImage(original_image, el[1]["image_path"], el[1]["label"])
```

```python
df = pd.DataFrame(df)
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
dataset = '../Data/5.augmented_dataset.csv'
df.to_csv(dataset, index=False)
```
